### Чтение файла, изучение общей информации. 

```python
# импорт необходимых библиотек
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
from scipy import stats as st
import scipy.stats
#считывание файла датасета
data = pd.read_csv('/datasets/games.csv')
# вывод 10 первых строк датасета
display(data.head(10))
# вывод общей информации по датасету
print(data.info())
# построение гистограмм
data.hist(figsize=(15, 20))
plt.show()
```

### Преобработка данных
#### Приведение названий столбцов к нижнему регистру

```python
# изменение название столбцов
data.columns = data.columns.str.lower()
# вывод изменённых названий столбцов
print(data.columns)
```

#### Подсчет пропусков 

```python
# смотрим количество пропущенных значений
print(data.isna().sum())
```

*Вывод по пропускам: Пропуски в столбцах 'critic_score', 'user_score', 'rating' некоректно заменять полностью каким то одним значением, на мой взгляд так как это исказить объективную оценку игр, пропуски в столбце 'year_of_release' думаю следует удалить, их не заменить значением из названий игр или какой либо другим(пример Fifa 2004 вышла в 2003, а не в 2004г.* 

```python
# удаление пропусков в годе релиза и преведение к нужному типу данных
data = data.dropna(subset = ['year_of_release'])
data['year_of_release'] = data['year_of_release'].astype('int')
print(data.isna().sum())
print(data.info())
```

```python
print(data[data['genre'].isna()])
```

* Пропуски в 'name' и 'genre' состоят из двух строк, в этих строках так же отсутсвует значение, рейтинга, оценки критиков и пользователей, данные строки лучше удалить


```python
# Удаление неинформативных строк
data = data.drop(labels = [659,14244],axis = 0)
# просмотр оставшихся пропусков
print(data.isna().sum())
```

*В столбце "user_score" присутствует значение 'tbd' расшифровывается как - "To Be Determined", то есть "Будет определено" можно было бы перевести пропуски в это значение или преобразовать их вместе с Nan в 0 или медиану, но нам во первых необходим числовой формат столбца для проверки гипотез плюс медиана и 0 исказить данные данные проверки,  поэтому преобразуем эти дополнительные пропуски данных в значение 'NaN'*

```python
# Подсчет дополнительных пропусков
print(len(data[data['user_score'] == 'tbd']))
# замена значений 'tbd' на 'NaN'
data.loc[data['user_score'] == 'tbd','user_score'] = float('nan')
# преобразуем данные столбца в нужный формат методом 'заглушки'
data['user_score'] = data['user_score'].fillna(-1)
data['user_score'] = data['user_score'].astype('float')
data.loc[data['user_score'] == -1, 'user_score'] = float('nan')
print(data.isna().sum())
```

Пропуски в столбце 'rating' я считаю можно заполнить заглушкой - 'not_defined'

```python
data['rating'] = data['rating'].fillna('not_defined')
print(data.isna().sum())
```

### Исследовательский анализ данных
####  Исследование, сколько игр выпускалось в разные годы

```python
# создание таблицы с расчетом количества выхода игр по годам
range_year=data['year_of_release'].value_counts().to_frame()
# Переименование столбцов и сортировка для лучшего отображения
range_year = range_year.rename( columns={'year_of_release':'Количество_игр'})
range_year = range_year.rename_axis('Год_релиза')
range_year = range_year.sort_values(by='Год_релиза')
# Вывод таблицы на экран
display(range_year)
# Вывод графика
range_year.plot(kind = 'bar', 
                        grid = True, 
                        figsize = (9, 5),
                        title='Количество игр выпущеный за период с 1980 по 2016г. за каждый год')
```

#### Исследование, как менялись продажи по платформам.

Поскольку на каждом этапе развития игровой индустрии продажи игр были существенно разные и неравномерные по количеству игр, некоректно расматривать общие продажи по всему временому периоду, для того чтобы отслежить *за какой характерный срок появляются новые и исчезают старые платформы* необходимо разбить датасет на периоды:
- Выделить лидирующие платформы за определенный период
- Отследить жизненный цикл лидирующих платформ

##### Статистика продаж самых популярных платформ за 1980-1993 год 

```python
# Создание таблицы с расчётом общих продаж по  за нужный год платфомам
stat_platform_1980 = data.loc[data['year_of_release'] < 1993].pivot_table( index = 'platform', 
                               values = 'sum_sales', 
                               aggfunc = sum).sort_values(by='sum_sales',ascending=False)
   
# Переименование столбцов для лучшей визуализации
stat_platform_1980 = stat_platform_1980.rename(columns = {'sum_sales':'Общие_продажи_млн'})
stat_platform_1980 = stat_platform_1980.rename_axis('Основные_платформы')
# Вывод получившийся таблицы на экран
display(stat_platform_1980.head(10))
stat_platform_1980.plot(kind = 'bar', 
                        grid = True, 
                        figsize = (9, 5),
                        title='Статистика продаж самых популярных платформ за 1980-1993 год')
```

*Делаем вывод что лидером этого цикла игровой индустрии была платформа NES(Nintendo Entertainment System) так у неё самые большие продажи*

##### Статистика продаж платформы NES((Nintendo Entertainment System))

```python
nes_stat= data.loc[data['platform'] == 'NES']
nes_stat=nes_stat.pivot_table(index='year_of_release', 
                              values = 'sum_sales', 
                              aggfunc = sum)
# Переименование столбцов и сортировка для лучшего отображения
nes_stat = nes_stat.rename(columns = {'sum_sales':'Общие_продажи_млн'})
nes_stat = nes_stat.rename_axis('Год')
nes_stat = nes_stat.sort_values(by='Год')
# Вывод таблицы на экран
display(nes_stat)
nes_stat.plot(y = 'Общие_продажи_млн', 
              kind = 'bar', 
              grid = True, 
              figsize = (9, 5),
              title='Статистика продаж платформы Nintendo Entertainment System(NES)')
```

Общее время продаж составило 11 лет, жизненный цикл это основное время продаж приставки составило 6 лет с 1984 по 1989 год, основные продажи (за исключением первого года выхода), затем продажи существено упали в 1989г. есть предположение что такому обвалу способствовал выход приставки следующего поколения, следующие 5 лет продажи непревышали 20 млн и в 1994 снизились почти до 0 и  на основе этого предлагаю иследовать следующий временной период с 1989 по 2000г


##### Статистика продаж самых популярных платформ за 1989-2001 год 

```python
# Создание таблицы с расчётом общих продаж по  за нужный год платфомам
stat_platform_1989 = data.loc[(data['year_of_release'] >= 1989) 
                              & (data['year_of_release'] <= 2001)].pivot_table( index = 'platform', 
                               values = 'sum_sales', 
                               aggfunc = sum).sort_values(by='sum_sales',ascending=False)
# Переименование столбцов для лучшей визуализации
stat_platform_1989 = stat_platform_1989.rename(columns = {'sum_sales':'Общие_продажи_млн'})
stat_platform_1989 = stat_platform_1989.rename_axis('Основные_платформы')
# Вывод получившийся таблицы и графика на экран 
display(stat_platform_1989.head(10))
stat_platform_1989.plot(kind = 'bar', 
                        grid = True, 
                        figsize = (9, 5),
                        title='Статистика продаж самых популярных платформ за 1989-2001 год')
```

С большим преобладанием лидирует PS её жизненный цикл и следует расмотреть

##### Исследование, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков.

Так как в целях исследования говорится о влияние отзывов **в одной популярной платформе** возьмем за пример самую популярную платформу в настоящий момент - PS4

```python
# Выделяем отдельную платформу
ps4_data = real_data.loc[real_data['platform'] == 'PS4'].dropna()
# Создаем таблицу распределение оценок пользователей по продажам
ps4_user_pivot = real_data.loc[real_data['platform'] == 'PS4'].pivot_table(index = 'user_score',
                                                                           values='sum_sales',
                                                                           aggfunc ='median').sort_values(by='user_score',
                                                                                                       ascending=False)
# Создаем таблицу распределение оценок критиков по продажам                                                                                                      
ps4_critic_pivot = real_data.loc[real_data['platform'] == 'PS4'].pivot_table(index = 'critic_score',
                                                                             values='sum_sales',aggfunc ='median').sort_values(by='critic_score',
                                                                                                                               ascending=False)
# Переименование столбцов для лучшей визуализации
ps4_user_pivot = ps4_user_pivot.rename(columns = {'sum_sales':'Общие_продажи_млн'})
ps4_user_pivot = ps4_user_pivot.rename_axis('Оценка_пользователей')
ps4_critic_pivot = ps4_critic_pivot.rename(columns = {'sum_sales':'Общие_продажи_млн'})
ps4_critic_pivot = ps4_critic_pivot.rename_axis('Оценка_критиков')

# Расчет корреляции
user_corr = ps4_data['sum_sales'].corr(ps4_data['user_score']).round(3)
critic_corr = ps4_data['sum_sales'].corr(ps4_data['critic_score']).round(3)

# Вывод статистики на экран
display(ps4_user_pivot)
display(ps4_critic_pivot)
print(f'Корреляция оценок пользователей и суммарных продаж равна {user_corr}')
print(f'Корреляция оценок критиков и суммарных продаж равна {critic_corr}')

# Вывод диаграмм рассеивания на экран
ps4_data.plot(x = 'user_score', 
              y = 'sum_sales',
              xlabel = 'Оценки пользователей',
              ylabel = 'Продажи(млн.)',
              style = 'o',
              figsize=(14,6),
              alpha=0.9, 
              title='Диаграмма рассеивания по продажам и оценкам пользователей',
              c = 'r')

ps4_data.plot(x = 'critic_score',
              y = 'sum_sales',
              xlabel = 'Оценки критиков',
              ylabel = 'Продажи(млн.)',
              style = 'o',
              figsize=(14,6),
              alpha=0.9,
              title='Диаграмма рассеивания по продажам и оценкам критиков',
              c = 'g')
plt.show
```

Вывод: Корреляции между продажами и оценками пользователей абсолютно нет, между продажами и оценками критиков есть но слабая, для проверки можно расчитать корреляцию так же на менее популярных платформах в лице XOne и 3DS

```python
# Выделяем данные под отдельную платформу
xone_data = real_data.loc[real_data['platform'] == 'XOne']
data_3ds = real_data.loc[real_data['platform'] == '3DS']
# Расчет корреляции xone
user_corr_xone = xone_data['sum_sales'].corr(xone_data['user_score']).round(3)
critic_corr_xone = xone_data['sum_sales'].corr(xone_data['critic_score']).round(3)
# Расчет корреляции 3ds
user_corr_3ds = data_3ds['sum_sales'].corr(data_3ds['user_score']).round(3)
critic_corr_3ds = data_3ds['sum_sales'].corr(data_3ds['critic_score']).round(3)
# Вывод корреляции на экран
print(f'Корреляция оценок пользователей и суммарных продаж платформы XOne равна {user_corr_xone}')
print(f'Корреляция оценок критиков и суммарных продаж платформы XOne  равна {critic_corr_xone}')
print(f'Корреляция оценок пользователей и суммарных продаж платформы 3DS  равна {user_corr_3ds}')
print(f'Корреляция оценок критиков и суммарных продаж платформы 3DS  равна {critic_corr_3ds}')
```

Как видно корреляция по отзывам по платформам в целом схожая и можно сделать вывод что между пользовательскими оценками и продажами нет корреляции, между оценками критиков и продажами есть но слабая

### Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

 *Поскольку мы проверяем равенство двух независимых выборок в один период времении в исследовании мы применим t-критерий Стьюдента  для проверки равенства средних двух генеральных совокупностей).
  

  Так как нулевая гипотеза обычно предпалагает равенство двух совокупностей, а альтернативная ей противоположное то:
- Нулевая гипотеза - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые*
- Альтернативная гипотеза - Средние пользовательские рейтинги платформ Xbox One и PC разное(в большую или меньшую сторону)*

```python
# Создаем массив как совокупность пользовательских рейтингов платформы XOne( Xbox One)
xone_user_score= real_data.loc[real_data['platform'] == 'XOne']['user_score'].dropna().to_numpy()
# Совокупность пользовательских рейтингов платформы PC
pc_user_score= real_data.loc[real_data['platform'] == 'PC']['user_score'].dropna().to_numpy()
# Критический уровень статистической значимости
alpha = 0.05 
# Применяем тест проверки равенства двух генеральных совокупностей
results = st.ttest_ind(xone_user_score, pc_user_score, equal_var = False)
print('p-значение:', results.pvalue)
# Вывод результата
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, средние пользовательские рейтинги разные")
else:
    print("Не получилось отвергнуть нулевую гипотезу, средние пользовательские рейтинги одинаковые ")




```

#### Средние пользовательские рейтинги жанров Action  и Sports разные

  Так как нулевая гипотеза обычно предпалагает равенство двух совокупностей, а альтернативная ей противоположное то:
- Нулевая гипотеза - Средние пользовательские рейтинги жанров Action и Sports одинаковые
- Альтернативная гипотеза - Средние пользовательские рейтинги жанров Action и Sports разные*

```python
# Совокупность пользовательских рейтингов жанра Action
action_user_score= real_data.loc[real_data['genre'] == 'Action']['user_score'].dropna().to_numpy()
# Совокупность пользовательских рейтингов жанра Sports
sports_user_score= real_data.loc[real_data['genre'] == 'Sports']['user_score'].dropna().to_numpy()
# Критический уровень статистической значимости
alpha = 0.05 
# Применяем тест проверки равенства двух генеральных совокупностей
results = st.ttest_ind(action_user_score, sports_user_score, equal_var = False)
print('p-значение:', results.pvalue)
# Вывод результата
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, средние пользовательские рейтинги разные")
else:
    print("Не получилось отвергнуть нулевую гипотезу, средние пользовательские одинаковые ")
```
